```python
%matplotlib inline
from __future__ import (print_function,
                        unicode_literals,
                        division)
from future.builtins import str, open, range, dict
import matplotlib
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
import sys
from matplotlib import gridspec

def plot_snap(start_t, end_t, volts, spikes_in, spikes_out, ts=1.0, start_neuron=0, end_neuron=-1, 
              crop_pre_spikes=False, base_value=None):
    ms_to_ts = 1.0/ts
    ts_to_ms = 1.0/ms_to_ts
    ts_start = int( start_t * ms_to_ts )
    ts_end = int( end_t * ms_to_ts )
    start_nid = start_neuron
    end_nid = end_neuron

    post_spikes = []
    post_ids = []
    for i in range(start_nid, end_nid):
        local_spikes = np.array(spikes_out[i])
        whr = np.where(np.logical_and(
                start_t <= local_spikes,
                local_spikes < end_t))
        post_spikes.append(local_spikes[whr])
        if len(whr[0]):
            post_ids.append(i)
    print(post_ids)

    pre_spikes = []
    if crop_pre_spikes:
        for i in range(start_nid, end_nid):
            local_spikes = np.array(spikes_in[i])
            whr = np.where(np.logical_and(
                    start_t <= local_spikes,
                    local_spikes < end_t))
            pre_spikes.append(local_spikes[whr])
    else:
        for i in range(len(spikes_in)):
            local_spikes = np.array(spikes_in[i])
            whr = np.where(np.logical_and(
                    start_t <= local_spikes,
                    local_spikes < end_t))
            pre_spikes.append(local_spikes[whr])

    plt.figure(figsize=(12, 8))
    ax = plt.subplot(1, 1, 1)

    for i, times in enumerate(pre_spikes):
        for t in times:
            plt.axvline((t - ts_to_ms - start_t)*ms_to_ts, linestyle='--', linewidth=0.5, color='magenta')

    for i, times in enumerate(post_spikes):
        for t in times:
            plt.axvline((t - ts_to_ms - start_t)*ms_to_ts, marker='.', linestyle='--', linewidth=0.5, color='blue')
    
    for i in range(start_nid, end_nid):
        plt.plot(volts[ts_start:ts_end, i], label=str(start_nid + i))
    if base_value is not None:
        plt.axhline(base_value, color='black')
    xticks = ax.get_xticks()
    ax.set_xticklabels(start_t + np.array(xticks)*ts_to_ms)
    ax.set_xlabel('time [ms]')
    ax.set_ylabel('membrane voltage [mV]')
    
    ax.legend()
    
    plt.show()



def plot_in_range(fig, ax, spikes, start_t, end_t, start_neuron=0, end_neuron=-1,
                  color='blue', markersize=1, marker='.', markeredgewidth=0, 
                  plot_line=None, linewidth=0, linestyle='-', alpha=1.0, label=None):
    
    first_plot = True
    for neuron_id, times in enumerate(spikes[start_neuron:end_neuron]):
        neuron_id += start_neuron
        times = np.array(times)
        whr = np.where(np.logical_and(times >= start_t,  times <= end_t))[0].astype('int')
        if len(whr):
            plot_times = times[whr]
            if plot_line is not None:
                for t in plot_times:
                    in_label = label if first_plot else None
                    first_plot = False

                    if plot_line == 'vertical':
                        plt.axvline(t, color=color, linewidth=linewidth, linestyle=linestyle, alpha=alpha,
                                    label=in_label)
                    elif plot_line == 'horizontal':
                        plt.axhline(t, color=color, linewidth=linewidth, linestyle=linestyle, alpha=alpha,
                                    label=in_label)
            else:
                in_label = label if first_plot else None
                first_plot = False
                plt.plot(plot_times, neuron_id*np.ones_like(plot_times), marker, color=color, alpha=alpha, 
                         markersize=markersize, markeredgewidth=markeredgewidth, label=in_label)


def active_per_pattern(spikes, dt=50):
    end = 0
    n_active = []
    
    for start in range(0, total_t, dt):
        sys.stdout.write("\r{}/{}".format(start, total_t))
        sys.stdout.flush()
        end = start + dt
        active_count = 0
        for times in spikes:
            
            times = np.array(times)
            whr = np.where(np.logical_and(times >= start, times < end))[0]
            if len(whr):
                active_count += 1

        n_active.append(active_count)
    
    sys.stdout.write("\n")
    sys.stdout.flush()
    
    return n_active


            
def pop_rate_per_second(spikes, total_t, dt):
    ms_to_s = 1./1000.
    n_neurons = len(spikes)
    end = 0
    rates = []
    for start in range(0, total_t, dt):
        end = start + dt
        spike_count = 0
        for times in spikes:
            times = np.array(times)
            whr = np.where(np.logical_and(times >= start, times < end))[0]
            spike_count += len(whr)
        rate = float(spike_count) / float(n_neurons * dt * ms_to_s)
#         rate = float(spike_count) / float(dt * ms_to_s)
        rates.append(rate)

    return rates    


def active_neurons_per_pattern(spikes, start_time, end_time, sample_indices, start_indices_index, config):
    t_per_sample = config['time_per_sample']
    n_samples = config['n_samples']
    n_patterns = config['n_patterns']
    
    posts = []
    active_neurons = {idx: set() for idx in range(n_patterns)}
    st = start_time
    for idx in range(start_indices_index, len(sample_indices)):
        pat_id = sample_indices[idx] // n_samples
        et = st + t_per_sample
        posts[:] = []
        for post_id, times in enumerate(spikes):
            times = np.array(times)
            find = np.where(np.logical_and(times >= st, times < et))[0]
            if len(find):
                posts.append(post_id)
        
#         print(pat_id, st, et, posts)
        ### update set of active neurons
        active_neurons[pat_id] |= set(posts)
        
        st = et

    return active_neurons


def avg_mean(sig):
    return np.sum(sig)/float(len(sig))

def energy(sig):
    return np.sum(np.power(sig, 2))

def power(sig):
    return energy(sig)/float(len(sig))

def variance(sig):
    return power(sig) - avg_mean(sig)**2

```

```python
def plot_weights(start_t, end_t, weights, spikes_in, spikes_out, n_post, start_neuron, end_neuron, ts=1.0, sim_step=0.1):
    markers = [k for k in matplotlib.markers.MarkerStyle.markers if not (k is None or k == 'None' or k=='' or k==' ')]
    n_markers = len(markers)
    ms_to_ts = 1.0/ts
    ts_to_ms = 1.0/ms_to_ts
    ts_start = int( start_t * ms_to_ts ) 
    ts_end = int( end_t * ms_to_ts )+1
    start_nid = start_neuron
    end_nid = end_neuron

    print(ms_to_ts, ts_to_ms, ts_start, ts_end, start_nid, end_nid)
    
    post_ids = []
    post_spikes = []
    for i in range(start_nid, end_nid):
        local_spikes = np.array(spikes_out[i])
        whr = np.where(np.logical_and(
                start_t <= local_spikes,
                local_spikes < end_t))
        if len(whr[0]):
            post_ids.append(i)
        post_spikes.append(local_spikes[whr] - sim_step)
    
    pre_ids = []
    pre_spikes = []
    for i in range(len(spikes_in)):
        local_spikes = np.array(spikes_in[i])
        whr = np.where(np.logical_and(
                start_t <= local_spikes,
                local_spikes < end_t))
        if len(whr[0]):
            pre_ids.append(i)
        pre_spikes.append(local_spikes[whr] - sim_step)


    plt.figure(figsize=(12, 8))
    ax = plt.subplot(1, 1, 1)
#     plt.axhline(0, linestyle=')
    for i, times in enumerate(pre_spikes):
        for t in times:
#             print("pre {} spiked at {}".format(i, times))
            plt.axvline(t, linestyle='--', linewidth=0.5, color='magenta')

    for i, times in enumerate(post_spikes):
        for t in times:
#             print("post {} spiked at {}".format(i, times))
            plt.axvline(t, linestyle='--', linewidth=0.5, color='blue')

    print("pre spiked ", pre_ids)
    print("post spiked ", post_ids)

            
    w_ids = []
    
    for pre in pre_ids:
        for post in post_ids:
            w_ids.append(pre*n_post + post)
    print("which should have changed from spike times", len(w_ids))

#     print(weights[0, :].reshape(-1, n_post).shape)
    
    epsilon = 1e-5
#     wdiff = weights[1,:] - weights[0, :]
#     print("raw abs where ", np.where(np.abs(wdiff) >= epsilon))
#     changed = np.random.choice(np.arange(weights.shape[1]), size=500, replace=False)
#     changed = np.arange(5000)
    changed = np.empty((1))
    for ts in range(ts_start+1, ts_end):
        chg = np.where(np.abs(weights[ts,:] - weights[ts-1, :]) > epsilon)[0]
        changed = np.union1d(changed, chg)
#     changed = w_ids
#     print((np.abs(wdiff)[changed]).sum())
#     print("num weight changed ", changed.size)
#     print("weight ids ", changed)
#     print("weight post ", changed%n_post)
#     print("weight pre ", changed//n_post)
    inter = np.intersect1d(changed, w_ids)
    print("change match ", inter.size)
#     print()

#     for i in w_ids:
#         print()
#         print(weights[ts_start:ts_end, i])
    print("which changed", changed.size)
    for i in changed:
        diff0 = np.zeros_like(weights[ts_start:ts_end, i])
        for j, ti in enumerate(range(ts_start+1, ts_end)):
            d = weights[ti, i] -  weights[ti-1, i]
#             if np.abs(d) <= epsilon:
#                 diff0[ti] = diff0[ti-1]
#             else:
#                 diff0[ti] = d
            diff0[j] = d
        dy = (np.max(diff0) * 0.05) * np.random.uniform(-1., 1., size=diff0.size) * (np.abs(diff0) > epsilon)
        dx = np.random.uniform(-2., 2., size=diff0.size) * (np.abs(diff0) > epsilon)
        times = (np.arange(ts_start, ts_end, dtype='int'))*ts_to_ms
        whr = np.where(np.abs(diff0) > epsilon)
        plt.plot((times + dx)[whr],  (diff0 + dy)[whr], #markersize=2,
                 marker=markers[int(i)%n_markers], linestyle='none')
#         plt.plot((np.arange(ts_start, ts_end, dtype='int'))*ts_to_ms,  weights[ts_start:ts_end, i], 
#                  linewidth=2.0, alpha=0.3, marker='.')

#     changed = np.where(wdiff >= epsilon)[0]
#     print("num weight increased ", changed.size)
#     print("weight ids ", changed)
#     print("weight pre ", changed//n_post)
#     print("weight post ", changed%n_post)
#     print()
    
#     changed = np.where(wdiff <= -epsilon)[0]
#     print("num weight decreased ", changed.size)
#     print("weight ids ", changed)
#     print("weight pre ", changed//n_post)
#     print("weight post ", changed%n_post)
#     print()
    
#     plt.plot(weights[ts_start:ts_end, changed], linewidth=0.1)
    
#     ax.set_xlim(start_t, end_t)
    xticks = ax.get_xticks()
#     ax.set_xticklabels(start_t + np.array(xticks)*ts_to_ms)
    ax.set_xlabel('time [ms]')
    ax.set_ylabel('weight')
    ax.margins(0.1, 0.1)
    plt.grid()
    plt.show()

    
data = np.load(fname)
weights = data['weights']
out_spikes = data['decision_spikes']
k_spikes = data['kenyon_spikes']
n_post = data['args'].item().nDN

ts = data['timestep']
sample_ts = data['weight_sample_dt']
# start_t = 900.
# end_t = 950.
# start_nid = 10
# end_nid = 20

start_t = 50.
end_t = 125.
start_nid = 0
end_nid = 100
print(weights.shape, ts, n_post)
plot_weights(start_t, end_t, weights, k_spikes, out_spikes, n_post, start_nid, end_nid, ts=sample_ts, sim_step=ts)
```
