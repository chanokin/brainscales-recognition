```python
in_labels = data['in_labels'].item()
print(in_labels)
```

```python
in_spikes = data['in_spikes'].item()
count = 0
plt.figure()
for lyr in in_spikes:
    spk = in_spikes[lyr]
    for nid, times in enumerate(spk):
        plt.plot(times, np.ones_like(times)*nid + count, '.b', markersize=3)
    count += len(spk)
    plt.axhline(count, color='gray', linestyle='--', linewidth=1.)
plt.show()
```

```python
for lyr in sorted(in_spikes.keys()):
    s1 = in_spikes[lyr]
    w, h = (28, 28) if lyr < 2 else (28//5+1, 28//3+1)
    dt = 1#ms
    total_t = 15
    num_images = int(np.ceil(total_t/float(dt)))
    imgs = [np.zeros((h, w)) for _ in range(num_images)]
    for idx, times in enumerate(s1):
        for t in times:
            r, c = int(idx//w), int(idx%w)
            imgs[int(t // dt)][r, c] += 1.0
    fw = 2
    plt.figure(figsize=(total_t*fw, fw))
    for idx, img in enumerate(imgs):
        ax = plt.subplot(1, total_t, idx+1)
        plt.imshow(img)
    plt.show()
```
