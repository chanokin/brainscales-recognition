```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt



```

```python
delay_exp = np.load('delay_experiments.npz')
experiments = delay_exp['experiments'].item()
plt.figure()
ax = plt.subplot()
plt.axvline(0, linestyle='--', color='gray')
plt.axhline(0, linestyle='--', color='gray')

for delay in experiments:
    if delay != 1 and delay != 3:
        continue
    dt_dw = experiments[delay]
    dts = sorted(dt_dw.keys())
    dws = [dt_dw[dt] for dt in dts]
    plt.plot(dts, dws, label='delay = %u'%delay)

max_dw = np.max(np.abs(dws)) * 1.5
ax.set_ylim(-max_dw, max_dw)
ax.set_xlabel(r'$\Delta t = t_{pre} - t_{post}$ [ms]')
ax.set_ylabel(r'$\Delta w $')
plt.legend()
plt.grid()
plt.show()

```

```python
compound_exp = np.load('compound_synapse.npz')
print(compound_exp.keys())
dt_dw = compound_exp['dt_dw'].item()
total_dt_dw = compound_exp['total_dt_dw'].item()


plt.figure()#figsize=(10, 8))
ax = plt.subplot()
plt.axvline(0, linestyle='--', color='gray')
plt.axhline(0, linestyle='--', color='gray')

for d in dt_dw:
    dts = sorted(dt_dw[d].keys())
    dws = [dt_dw[d][dt] for dt in dts]
    label = 'standard' if d == 1 else 'inverse'
    plt.plot(dts, dws, label=label, linewidth=1.0, linestyle='--')
        

dts = sorted(total_dt_dw.keys())
dws = [total_dt_dw[dt] for dt in dts]
plt.plot(dts, dws, label='compound', linewidth=2.0)



max_dw = np.max(np.abs(dws)) * 2.5
ax.set_ylim(-max_dw, max_dw)
ax.set_xlabel(r'$\Delta t = t_{pre} - t_{post}$ [ms]')
ax.set_ylabel(r'$\Delta w $')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('compound_stdp.pdf')
plt.show()
```

```python
end_w = compound_exp['end_w'].item()

plt.figure()#figsize=(10, 8))
ax = plt.subplot()
plt.axvline(0, linestyle='--', color='gray')
plt.axhline(0.005, linestyle='--', color='gray')

for d in end_w:
    dts = sorted(end_w[d].keys())
    ws = [end_w[d][dt] for dt in dts]
    plt.plot(dts, ws, label='delay = %u'%(d), linewidth=1.0, linestyle='--')


# max_dw = np.max(np.abs(dws)) * 2.5
# ax.set_ylim(-max_dw, max_dw)
ax.set_xlabel(r'$\Delta t = t_{pre} - t_{post}$ [ms]')
ax.set_ylabel(r'$w$')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('compound_stdp.pdf')
plt.show()
```

```python
end_w = compound_exp['end_w'].item()

plt.figure()#figsize=(10, 8))
ax = plt.subplot()
plt.axvline(0, linestyle='--', color='gray')
plt.axhline(0.010, linestyle='--', color='gray')

for d in end_w:
    dts = sorted(end_w[d].keys())
    ws = [end_w[d][dt] for dt in dts]
        
sum_w = {dt: 0.0 for dt in dts}
for d in end_w:
    for dt in end_w[d]:
        sum_w[dt] += end_w[d][dt]

dts = sorted(total_dt_dw.keys())
ws = [sum_w[dt] for dt in dts]
plt.plot(dts, ws, label='compound', linewidth=2.0)



# max_dw = np.max(np.abs(dws)) * 2.5
# ax.set_ylim(-max_dw, max_dw)
ax.set_xlabel(r'$\Delta t = t_{pre} - t_{post}$ [ms]')
ax.set_ylabel(r'$w$')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('compound_stdp.pdf')
plt.show()
```
